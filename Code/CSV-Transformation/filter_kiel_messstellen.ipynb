```python
import polars as pl
import os
```

```python
# Liste der ausgewählten Dauerzählstellen
messstellen = [1116, 1111, 1112, 1158, 1135, 1156, 1104, 1162, 1194]

BASE_DIR = "../../data"
OUTPUT_DIR = os.path.join(BASE_DIR, "BASt", "AggregatedDataForKiel")

# CSV-Datei im Lazy-Modus einlesen, nach den ausgewählten Messstellen filtern und das Ergebnis als DataFrame in den Speicher laden
df_filtered = (pl.scan_csv(os.path.join(BASE_DIR, "BASt", "AggregatedDataForKiel", "kiel_stations_raw.csv")).filter(pl.col("Zst").is_in(messstellen)).collect())

# Gefilterte Daten als neue CSV-Datei speichern
df_filtered.write_csv(os.path.join(OUTPUT_DIR, "kiel_specific_stations_raw.csv"))
```
