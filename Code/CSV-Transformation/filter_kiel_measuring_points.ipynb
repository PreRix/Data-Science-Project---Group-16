```python
import polars as pl
import os
```

```python
# List of the selected measuring points
measuring_points = [1116, 1111, 1112, 1158, 1135, 1156, 1104, 1194]

BASE_DIR = "../../data"
OUTPUT_DIR = os.path.join(BASE_DIR, "BASt", "AggregatedDataForKiel")

# read .csv file and filter for selected measuring points
df_filtered = (pl.scan_csv(os.path.join(BASE_DIR, "BASt", "AggregatedDataForKiel", "kiel_stations_raw.csv")).filter(pl.col("Zst").is_in(measuring_points)).collect())

# Save filtered data in .csv
df_filtered.write_csv(os.path.join(OUTPUT_DIR, "kiel_specific_stations_raw.csv"))
```
